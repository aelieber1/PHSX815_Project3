```python
# importing poisson from scipy
from scipy.stats import poisson
 
# importing numpy as np
import numpy as np
 
# importing matplotlib as plt
import matplotlib.pyplot as plt
 
 
# creating a numpy array for x-axis
# using step size as 1
x = np.arange(0,15, 0.25)
 
# poisson distribution data for y-axis
y = poisson.pmf(x, mu=3, loc=0)
 
 
# plotting the graph
plt.plot(x, y, color='indianred')
plt.title("Poisson Distribution for True Rate Value of $\lambda$ = 3")
plt.xlabel("Rate Values")
plt.ylabel("Probabilities")
 
# showing the graph
plt.show()
```
